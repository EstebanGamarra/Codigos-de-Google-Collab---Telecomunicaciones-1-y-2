1) [MICROONDAS - REFLEXIÓN EN EL SUELO] [50%] Supongamos un vano de radioenlace sobre esteros, con una longitud de $d= 65$ km y alturas de antenas sobre el nivel del mar de  $h_t=145$ m  y $h_r=360$ m en la frecuencia de 12 GHz.

El coeficiente de reflexión es 0,79 según se ha podido estimar, $\beta=\pi$ rad y las antenas no presentan discriminación directiva. Calcule para valores de k 4/3, 1,  4/5 y 2/3:


a.  El punto de reflexión.

b. El ángulo de incidencia.

c. El Factor de Divergencia.

d.  El retardo del rayo reflejado.

e. La frecuencia de máximo desvanecimiento.

f.  La máxima profundidad de desvanecimiento.

Analice luego la selectividad en frecuencia del efecto de la reflexión en el suelo, atendiendo un canal de 55 MHz.


```python
import math
import numpy as np

f=12e9
d=60
ht=145
hr=360
BW=55e6
f_ll=f-BW/2
f_ul=f+BW/2
Rmag=0.79
Beta=np.pi
 
gTR=1
gRR=1

print("Valores de K utilizados")
k=np.array([4/3, 1, 4/5, 2/3])
print(" k=",np.round(k,4))
ae=6370

#Distancia del transmisor al punto de reflexión, d1
p=2/math.sqrt(3)*np.sqrt(6.37 * k *(ht+hr) + (d/2)**2)
#print("p=",np.round(p,4))
#print(12.74*k*(hr-ht)*d,np.power(p,3))
phi=np.arccos(np.divide((12.74*k*(hr-ht)*d),np.power(p,3)))
#print("phi=",np.round(phi,4))
d2=d/2 + p*np.cos((np.pi+phi)/3)
d1=d-d2
print("a. El punto de reflexión está a :")
print(" d1=",np.round(d1,4), "km de la antena 1.")

#Angulo de incidencia
htp=ht-np.divide((4*d1**2),(51*k))
#print("htp=",np.round(htp,2))
hrp=hr-np.divide((4*d2**2),(51*k))
psi= np.rad2deg(((htp+hrp)/d)/1000)
print("b. El ángulo de incidencia es: ")
print(" psi=",np.round(psi,4), " grados")
 
#Factor de divergencia
D=1/np.sqrt((1+((5/(16*k)*(d1**2*d2/(d*htp))))))
print("c. El Factor de Divergencia es:")
print(" D=",np.round(D,4))

#Retardo del rayo reflejado
deltaL=(2*htp*hrp*1e-3)/d
tau=deltaL/300000000  #nanosegundos
print("d. El retardo del rayo reflejado es:")
print(" tau=",np.round(tau*1e9,4), " ns")

#FR
b=Rmag*D*gTR*gRR
FR_max=-20*np.log10(1-b)
 
c=Beta/np.pi
n0=np.floor(f*tau+c/2)
f0=np.divide((2*n0),(2*tau))
print("e. La frecuencia de máximo desvanecimiento es:")
print(" fmax_desvanecimiento=",np.round(f0*1e-9,4), "GHz")

print("f. La máxima profundidad del desvanecimiento, F_R(f)_max, es:")
print(" F_R(f)_max=",np.round(FR_max,4), " dB")

print("2.Para ver la selectividad en frecuencia analizamos la atenuación en los extremos del canal con el valor de k más exigente (2/3).")
fcanal=np.array([f-BW/2,f,f+BW/2])
print(" f=",fcanal, "Hz")
FR=-10*np.log10(1+b[3]**+2*b[3]*np.cos(2*np.pi*fcanal*tau[3]+Beta))
print(" FR(f)=",FR, " dB")
print("Conclusión: Del resultado se puede ver que el desvaneciemiento es ligeramente (poco) selectivo en frecuencia.")
```
