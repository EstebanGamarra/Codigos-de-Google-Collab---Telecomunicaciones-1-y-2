# 2. Calcule la ganancia a 6 GHz de un paraboloide de revolución de 2 m de diámetro con una eficiencia del 55%.

```python
import numpy as np
import math 

D = 2
eficiencia = 0.55
superficie = np.pi*(D**2)/4
f = 6e9
long_onda = 3e8/f

area = eficiencia*superficie


g = ((4*np.pi)/(long_onda**2))*area
G = 10* math.log10(g)

print("La ganancia es de: ",G)
```
