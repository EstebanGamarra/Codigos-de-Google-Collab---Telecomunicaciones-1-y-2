1) [MICROONDAS - REFLEXIÓN EN EL SUELO] [50%] Supongamos un vano de radioenlace sobre esteros, con una longitud de $d= 65$ km y alturas de antenas sobre el nivel del mar de  $h_t=145$ m  y $h_r=360$ m en la frecuencia de 12 GHz.

El coeficiente de reflexión es 0,79 según se ha podido estimar, $\beta=\pi$ rad y las antenas no presentan discriminación directiva. Calcule para valores de k 4/3, 1,  4/5 y 2/3:


a.  El punto de reflexión.

b. El ángulo de incidencia.

c. El Factor de Divergencia.

d.  El retardo del rayo reflejado.

e. La frecuencia de máximo desvanecimiento.

f.  La máxima profundidad de desvanecimiento.

Analice luego la selectividad en frecuencia del efecto de la reflexión en el suelo, atendiendo un canal de 55 MHz.


```python
import numpy as np
import math

print("MICROONDAS 1")
d = 65
ht = 145 
hr = 360
frec = 12
coefref = 0.79
k1 = 4/3
k2 = 1
k3 = 4/5
k4 = 2/3

R = 6370
ae1 = k1*R
ae2 = k2*R
ae3 = k3*R
ae4 = k4*R

# El ángulo de incidencia.

htprima1 = ht - (d**2/(2*ae1))
hrprima1 = hr - (d**2/(2*ae1))

htprima2 = ht - (d**2/(2*ae2))
hrprima2 = hr -(d**2/(2*ae2))

htprima3 = ht - (d**2/(2*ae3))
hrprima3 = hr - (d**2/(2*ae3))

htprima4 = ht - (d**2/(2*ae4))
hrprima4 = hr - (d**2/(2*ae4))

print("ITEM B")

#en mili-radianes
psi1 = (htprima1+hrprima1)/(d*1000)
psi2 = (htprima2+hrprima2)/(d*1000)
psi3 = (htprima3+hrprima3)/(d*1000)
psi4 = (htprima3+hrprima4)/(d*1000)

print("El punto de reflexion para k=4/3 es:",np.rad2deg(psi1),"º grados")
print("El punto de reflexion para k=1 es:",np.rad2deg(psi2),"º grados")
print("El punto de reflexion para k=4/5 es:",np.rad2deg(psi3),"º grados")
print("El punto de reflexion para k=2/3 es:",np.rad2deg(psi4),"º grados")

# El Factor de Divergencia.
print("ITEM C")
m1 = ((d**2)*1e3)/(4*ae1*(ht+hr))
m2 = ((d**2)*1e3)/(4*ae2*(ht+hr))
m3 = ((d**2)*1e3)/(4*ae3*(ht+hr))
m4 = ((d**2)*1e3)/(4*ae4*(ht+hr))

c = (ht-hr)/(ht+hr)

b1 = 2*np.sqrt((m1+1)/(3*m1))
b1arcos =  np.arccos((3*c*np.sqrt((3*m1)/((m1+1)**3))/2))
b1cos = np.cos(np.pi/3 + (1/3)*b1arcos)
b1 = b1*b1cos


b2 = 2*np.sqrt((m2+1)/(3*m2))
b2arcos =  np.arccos((3*c*np.sqrt((3*m2)/((m2+1)**3))/2))
b2cos = np.cos(np.pi/3 + (1/3)*b2arcos)
b2 = b2*b2cos

b3 = 2*np.sqrt((m3+1)/(3*m3))
b3arcos =  np.arccos((3*c*np.sqrt((3*m3)/((m3+1)**3))/2))
b3cos = np.cos(np.pi/3 + (1/3)*b3arcos)
b3 = b3*b3cos

b4 = 2*np.sqrt((m4+1)/(3*m4))
b4arcos =  np.arccos((3*c*np.sqrt((3*m4)/((m4+1)**3))/2))
b4cos = np.cos(np.pi/3 + (1/3)*b4arcos)
b4 = b4*b4cos

D1 = np.sqrt((1-m1*(1+b1**2))/(1+m1*(1-3*(b1**2))))
D2 = np.sqrt((1-m2*(1+b2**2))/(1+m2*(1-3*(b2**2))))
D3 = np.sqrt((1-m3*(1+b3**2))/(1+m3*(1-3*(b3**2))))
D4 = np.sqrt((1-m4*(1+b4**2))/(1+m4*(1-3*(b4**2))))

print("El factor de divergencia para k=4/3 es:",D1)
print("El factor de divergencia para k=1 es:",D2)
print("El factor de divergencia para k=4/5 es:",D3)
print("El factor de divergencia para k=2/3 es:",D4)

# El punto de reflexión.
print("ITEM A")

d1 = d*(1+b1)/2
d2 = d*(1+b2)/2
d3 = d*(1+b3)/2
d4 = d*(1+b4)/2

#ya que d1 y d2 se calculan para ht > hr pero tenemos hr < ht, las distancias se invierten para este cálculo

print("El punto de reflexión para k=4/3 es:",d1,"km desde la antena mas chica")
print("El  punto de reflexión para k=1 es:",d2,"km desde la antena mas chica")
print("El  punto de reflexión para k=4/5 es:",d3,"km desde la antena mas chica")
print("El  punto de reflexión para k=2/3 es:",d4,"km desde la antena mas chica")

# El retardo del rayo reflejado.
# PODEMOS CALCULARLO POR EL DESFASE DEL CAMPO REFLEJADO Y EL DIRECTO
print("ITEM D")

#en radianes (usamos km y frec en sus valores reales ya que trabajamos con m y km)

delta1 = (np.pi*frec*1e9*htprima1*hrprima1)/(75*d*1e3)
delta2 = (np.pi*frec*1e9*htprima2*hrprima2)/(75*d*1e3)
delta3 = (np.pi*frec*1e9*htprima3*hrprima3)/(75*d*1e3)
delta4 = (np.pi*frec*1e9*htprima4*hrprima4)/(75*d*1e3)

delta1div = delta1/(2*np.pi)
delta1 = delta1/delta1div

delta2div = delta2/(2*np.pi)
delta2 = delta2/delta2div

delta3div = delta3/(2*np.pi)
delta3 = delta3/delta3div

delta4div = delta4/(2*np.pi)
delta4 = delta4/delta4div

# ahora que sabemos que tiene un desfase de delta, sabemos que la onda recorre 
# por lo menos una longitud de onda proporcional al desfase
# de nuestra ecuacion original hallamos la longitud de onda original

lamb = 3e8/frec*1e9

#hallamos la relacion de proporcion

lamb1 = (lamb+delta1*lamb)/lamb
lamb2 = (lamb+delta2*lamb)/lamb
lamb3 = (lamb+delta3*lamb)/lamb
lamb4 = (lamb+delta4*lamb)/lamb

frec1 = 3e8/(lamb1*1e9)
frec2 = 3e8/(lamb2*1e9)
frec3 = 3e8/(lamb3*1e9)
frec4 = 3e8/(lamb4*1e9)

print("El retardo para k=4/3 es:",1+(lamb1 - 1)/(2*np.pi), "de la onda original")
print("El  retardo para k=1 es:",1+(lamb2 - 1)/(2*np.pi), "de la onda original")
print("El  retardo para k=4/5 es:",1+(lamb3 - 1)/(2*np.pi), "de la onda original")
print("El  retardo para k=2/3 es:",1+(lamb4 - 1)/(2*np.pi), "de la onda original")
#ahora vemos que 1/frec hallada es el retardo respecto a la frecuencia original

# La frecuencia de máximo desvanecimiento.
print("ITEM E")

# La máxima profundidad de desvanecimiento.
print("ITEM F")
# Selectividad en un canal de 55MHz
print("ITEM G")
```

2) [MICROONDAS - PRÁCTICO - CÁLCULO RADIOENLACE] [50%]  Una operada posee don torres  en la ruta 6 distancias 40 km. La primera es de 40 metros de altura y la segunda de 50 metros. Han verificado que el enlace (13 GHz) se ha degradado en los últimos meses, y notaron que en el vano se ha construido un silo de 30 metros de altura a un tercio del camino entre la primera y la segunda torre (en la ruta del rayo directo). ¿Se debe este problema al silo construido (intromisión en F$_1$)?

La altura del terreno es de 130 metros (primera torre), 132 metros (segunda torre) y 110 metros en el silo.

Justifique su respuesta.


```python
import numpy as np
import math

d = 40
h1 = 130+40
h2 = 132+50
frec = 13

hsilo = 110+30
d1 = (1/3)*d
d2 =(2/3)*d

#PREGUNTA ES, HUBO PERDIDAS, ES POR EL SILO
#PRIMERAMENTE HAY QUE HALLAR LA PERDIDA DE TRANSMISIÓN EN ESPACIO LIBRE:

Lbf = 92.4 + 20*np.log10(13) + 20*np.log10(d)

print("La pérdida de transmisión en espacio libre es de:",Lbf,"dB")

#AHORA HAY QUE HALLAR LA PERDIDA POR EL OBSTACULO MEDIANTE LA ELIPSOIDE DE FRESNEL

F1 = 17.3*np.sqrt((d1*d2)/(frec*d))

print("El radio primero de Fresnel es de:", F1, "metros")
print("La altura de la antena respecto de la primera antena es de:",h1-hsilo,"metros")

print("Conclusión: \n Ya que el obstaculo no se encuentra en el radio del primer elipsoide de Fresnel, no hay pérdidas por dicho obstáculo, (o son menores a 15 dB), \n en caso contrario un cálculo teórico asumiendo pérdidas mayores a (15 dB)... :")

#POR ESTE MOTIVO Y CONSIDERANDO QUE LA LINEA DE VISION DIRECTA ES LA DE DISTANCIA QUE EMPIEZA POR LA ALTURA DE LA PRIMERA ANTENA
#EL SILO NO CAUSA NINGUNA PERDIDA EN LA TRANSMISION

#SI EL SILO ESTUVIESE DENTRO DEL PRIMER RADIO DE FRESNEL:

#POR LA FORMULA DE DIFRACCION (CUANDO UNA ONDA RODEA UN OBSTACULO) 
#PARA HALLAR LA ALTURA MAXIMA HALLAMOS EL SENO (CATETO OPUESTO A LA ALTURA DEL OBSTACULO EN EL TRAYECTO DE LA LINEA DE VISIBILIDAD DIRECTA)
#SIN EMBARGO COMO LA DISTANCIA DEL CATETO ADYASCENTE ES DEMASIADO SUPERIOR A LA ALTURA DE LAS ANTENAS (CATETO OPUESTO)
#USAMOS LA DISTANCIA EN HORIZONTAL COMO LINEA DE VISIBILIDAD DIRECTA, ASI h = hobjeto - hantenamaschica

Lobs = (-20*30)/F1 +10
print("La pérdida de la señal con el obstáculo sería en teoría", Lbf-Lobs,"dB")

```
