$s(t) = A_c \cos (2 \pi f_c t + 2 \pi k_f \int_0^t m(\tau) d\tau)$

```python
import numpy as np
import matplotlib.pyplot as plt

#Carrier wave c(t)=A_c*cos(2*pi*f_c*t)
#Modulating wave m(t)=A_m*cos(2*pi*f_m*t)
#Modulated wave s(t)=A_c[1+mu*cos(2*pi*f_m*t)]cos(2*pi*f_c*t)

A_c = float (5) #(input('Enter carrier amplitude: '))
f_c = float (100) #(input('Enter carrier frquency: '))
k_f = float (100)
A_m = float (0.4) #(input('Enter message amplitude: '))
f_m = float (10) #(input('Enter message frquency: '))
#S_m = np.zeros (100,float ,'C')
#modulation_index = float (100) #(input('Enter modulation index: '))
t_s = 0.001
t = np.arange(0, 1, t_s)
#time = np.arange(0,1,100)

#print (t)
#print (S_m)

carrier = A_c*np.cos(2*np.pi*f_c*t)
modulator = A_m*np.cos(2*np.pi*f_m*t)
S_m = np.cumsum(modulator)*t_s
product = A_c*np.cos(2*np.pi*f_c*t + 2*np.pi*k_f*S_m)


plt.subplot(3,1,1)
plt.title('Frequency Modulation')
plt.plot(modulator,'g')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')

plt.subplot(3,1,2)
plt.plot(carrier, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(product, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('FM signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)

fig.savefig('Amplitude Modulation.png', dpi=100)
```
